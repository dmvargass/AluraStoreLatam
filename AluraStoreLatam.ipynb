### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
total_ingreso_tienda1 = tienda['Precio'].sum()
total_ingreso_tienda2 = tienda2['Precio'].sum()
total_ingreso_tienda3 = tienda3['Precio'].sum()
total_ingreso_tienda4 = tienda4['Precio'].sum()
```

```python
print(f"Ingreso total de la Tienda 1: {total_ingreso_tienda1}")
print(f"Ingreso total de la Tienda 2: {total_ingreso_tienda2}")
print(f"Ingreso total de la Tienda 3: {total_ingreso_tienda3}")
print(f"Ingreso total de la Tienda 4: {total_ingreso_tienda4}")
```

# 2. Ventas por categoría

```python
ventas_por_categoria_tienda1 = tienda['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda2 = tienda2['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda3 = tienda3['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda4 = tienda4['Categoría del Producto'].value_counts()
```

```python
print("Ventas por categoría en la Tienda 1:")
print(ventas_por_categoria_tienda1)
print("\nVentas por categoría en la Tienda 2:")
print(ventas_por_categoria_tienda2)
print("\nVentas por categoría en la Tienda 3:")
print(ventas_por_categoria_tienda3)
print("\nVentas por categoría en la Tienda 4:")
print(ventas_por_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()
```

```python
print(f"Calificación promedio de la Tienda 1: {calificacion_promedio_tienda1}")
print(f"Calificación promedio de la Tienda 2: {calificacion_promedio_tienda2}")
print(f"Calificación promedio de la Tienda 3: {calificacion_promedio_tienda3}")
print(f"Calificación promedio de la Tienda 4: {calificacion_promedio_tienda4}")
```

# 4. Productos más y menos vendidos

```python
conteo_productos_tienda1 = tienda['Producto'].value_counts()
top_productos_tienda1 = conteo_productos_tienda1.head()
bottom_productos_tienda1 = conteo_productos_tienda1.tail()

conteo_productos_tienda2 = tienda2['Producto'].value_counts()
top_productos_tienda2 = conteo_productos_tienda2.head()
bottom_productos_tienda2 = conteo_productos_tienda2.tail()

conteo_productos_tienda3 = tienda3['Producto'].value_counts()
top_productos_tienda3 = conteo_productos_tienda3.head()
bottom_productos_tienda3 = conteo_productos_tienda3.tail()

conteo_productos_tienda4 = tienda4['Producto'].value_counts()
top_productos_tienda4 = conteo_productos_tienda4.head()
bottom_productos_tienda4 = conteo_productos_tienda4.tail()
```

```python
print("Top 5 productos más vendidos en la Tienda 1:")
print(top_productos_tienda1)
print("\nTop 5 productos menos vendidos en la Tienda 1:")
print(bottom_productos_tienda1)

print("\nTop 5 productos más vendidos en la Tienda 2:")
print(top_productos_tienda2)
print("\nTop 5 productos menos vendidos en la Tienda 2:")
print(bottom_productos_tienda2)

print("\nTop 5 productos más vendidos en la Tienda 3:")
print(top_productos_tienda3)
print("\nTop 5 productos menos vendidos en la Tienda 3:")
print(bottom_productos_tienda3)

print("\nTop 5 productos más vendidos en la Tienda 4:")
print(top_productos_tienda4)
print("\nTop 5 productos menos vendidos en la Tienda 4:")
print(bottom_productos_tienda4)
```

# Visualizacion de resultados

```python
import matplotlib.pyplot as plt
import seaborn as sns

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [total_ingreso_tienda1, total_ingreso_tienda2, total_ingreso_tienda3, total_ingreso_tienda4]

plt.figure(figsize=(10, 6))
plt.plot(tiendas, ingresos, marker='o', linestyle='-', color='skyblue')
plt.title('Ingreso Total por Tienda (Gráfico de Líneas)')
plt.ylabel('Ingreso Total')
plt.xlabel('Tienda')
plt.grid(True)
plt.show()
```

```python
ventas_por_categoria_combinado = pd.DataFrame({
    'Tienda 1': ventas_por_categoria_tienda1,
    'Tienda 2': ventas_por_categoria_tienda2,
    'Tienda 3': ventas_por_categoria_tienda3,
    'Tienda 4': ventas_por_categoria_tienda4
})

ventas_por_categoria_combinado = ventas_por_categoria_combinado.fillna(0)

ventas_por_categoria_combinado.plot(kind='bar', figsize=(15, 8))
plt.title('Ventas por Categoría en Todas las Tiendas')
plt.xlabel('Categoría del Producto')
plt.ylabel('Número de Ventas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificaciones = [calificacion_promedio_tienda1, calificacion_promedio_tienda2, calificacion_promedio_tienda3, calificacion_promedio_tienda4]

plt.figure(figsize=(8, 8))
plt.pie(calificaciones, labels=tiendas, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('plasma', len(tiendas)))
plt.title('Proporción de Calificación Promedio por Tienda')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x=top_productos_tienda1.index, y=top_productos_tienda1.values, palette='cubehelix')
plt.title('Top 5 Productos Más Vendidos en la Tienda 1')
plt.xlabel('Producto')
plt.ylabel('Número de Ventas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
